```python
import requests
from bs4 import BeautifulSoup
import re
from textrank4zh import TextRank4Sentence
import os

news_dict = {'social': ['BBM54PGAwangning', '社会'], 'entertainment': ['BA10TA81wangning', '娱乐'], 'technology': ['BA8D4A3Rwangning', '科技'], 'military': ['BAI67OGGwangning', '军事'], 'travel': ['BEO4GINLwangning', '旅游']}


def creatdb(urlname, indexname):
    turl = requests.get('https://3g.163.com/touch/reconstruct/article/list/' + urlname + '/0-10.html').content.decode()
    urllist = re.findall('"url":"(.*?)"', turl)
    # print(urllist)
    rr = []
    index = indexname + '--'
    for i in urllist:
        if i and i.startswith('http'):
            conurl = requests.get(i)
            b = BeautifulSoup(conurl.text, 'lxml')
            textlist = set(b.find_all('p'))
            textresult = ''
            for j in textlist:
                textresult += str(j.text)
            tr4s = TextRank4Sentence()
            tr4s.analyze(text=textresult, source="no_stop_words")
            key_sentences = tr4s.get_key_sentences(2, sentence_min_len=2)
            result = '' + index + '、'
            for sentence in key_sentences:
                if sentence['sentence'] not in result:
                    result += sentence['sentence'] + '。 '
            print(result, '---->', textresult)
            print()
            if len(result) == 0:
                continue
            


for i in news_dict.items():
    creatdb(i[1][0], i[1][1])
```
